# Barchart Count

Bar chart of missing (or present) value counts per column.
`show_both=True` renders missing and present bars side by side.

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

```python
md.barchart_count().show()
```

## Intermediate

Horizontal orientation with value labels.

```python
md.barchart_count(
    title="Missing Count per Column",
    orientation="horizontal",
    show_values=True,
).show()
```

## Advanced

Missing and present bars side by side, dark theme, save.

```python
md.barchart_count(
    title="Missing vs. Present Count per Column",
    ignore_high_missingness=False,
    high_missingness_threshold=0.9,
    completeness_mode="least",
    completeness_threshold=0.9,
    max_columns_by_completeness=5,
    max_columns=15,
    order_by=[{"column": "__missing__", "ascending": False}],
    orientation="horizontal",
    show_values=True,
    value="missing",
    show_both=True,
    missing_color="#e74c3c",
    present_color="#2ecc71",
    background_color="#1a1a2e",
    text_color="#eaeaea",
    show_legend=True,
    legend_title="Data Status",
    width=900,
    height=500,
).save()
```
