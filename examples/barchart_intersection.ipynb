# Barchart Intersection

UpSet-style bar chart showing co-occurrence patterns: how many rows are simultaneously missing across specific combinations of columns.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

```python
md.barchart_intersection().show()
```

## Intermediate

Restrict to the three columns that actually have missing values.

```python
md.barchart_intersection(
    title="Missingness Intersection",
    selected_columns=["age", "deck", "embarked"],
    max_intersections=10,
).show()
```

## Advanced

Highlight `age`, custom dot/line sizes, light theme, save.

```python
md.barchart_intersection(
    title="Missingness Intersection — All Patterns",
    selected_columns=["age", "deck", "embarked"],
    max_sets=3,
    max_intersections=10,
    min_intersection_size=1,
    sort_order="desc",
    show_values=True,
    matrix_dot_size=14,
    matrix_line_width=4,
    excluded_dot_color="#cccccc",
    max_label_length=20,
    highlight_columns=["age"],
    highlight_color="#e377c2",
    missing_color="#e74c3c",
    present_color="#27ae60",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=True,
    legend_title="Count",
    width=800,
    height=500,
).save()
```
