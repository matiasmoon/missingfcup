# Barchart Rate

Bar chart of the missing rate per column, expressed as a fraction (0–1) or percentage.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

```python
md.barchart_rate().show()
```

## Intermediate

Percentage scale with value labels.

```python
md.barchart_rate(
    title="Missing Rate per Column",
    scale="percentage",
    show_values=True,
).show()
```

## Advanced

Horizontal orientation, custom colour, light theme, save.

```python
md.barchart_rate(
    title="Missing Rate — All Columns",
    ignore_high_missingness=False,
    high_missingness_threshold=0.9,
    completeness_mode=None,
    completeness_threshold=0.0,
    max_columns_by_completeness=0,
    max_columns=15,
    order_by=[{"column": "__missing__", "ascending": False}],
    orientation="horizontal",
    show_values=True,
    scale="percentage",
    missing_color="#9b59b6",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=True,
    legend_title="Missing Rate",
    width=900,
    height=450,
).save()
```
