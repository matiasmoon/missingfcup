# Barchart Total

Single bar chart showing total present vs. missing cell counts across the entire dataset — a quick overall completeness summary.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

```python
md.barchart_total().show()
```

## Intermediate

Add a title and legend label.

```python
md.barchart_total(
    title="Titanic: Overall Data Completeness",
    show_legend=True,
    legend_title="Data Status",
).show()
```

## Advanced

Custom colours, dark theme, save.

```python
md.barchart_total(
    title="Titanic: Overall Data Completeness",
    missing_color="#e74c3c",
    present_color="#2ecc71",
    background_color="#1a1a2e",
    text_color="#eaeaea",
    show_legend=True,
    legend_title="Data Status",
    width=700,
    height=400,
).save()
```
