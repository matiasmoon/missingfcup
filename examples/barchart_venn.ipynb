# Barchart Venn

Bar chart of the 7 exclusive missingness subsets for exactly 3 selected columns.
Each bar is one region of the Venn diagram: only A, only B, only C, A∩B, A∩C, B∩C, or all three.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

`age`, `deck`, and `embarked` are the three columns with any missing values in this dataset.

```python
md.barchart_venn(
    selected_columns=["age", "deck", "embarked"],
).show()
```

## Intermediate

Show percentages instead of counts with value labels.

```python
md.barchart_venn(
    selected_columns=["age", "deck", "embarked"],
    title="Missingness Subsets: age, deck, embarked",
    value="percent",
    show_values=True,
).show()
```

## Advanced

Ascending sort, custom colour, light theme, save.

```python
md.barchart_venn(
    selected_columns=["age", "deck", "embarked"],
    sort_order="asc",
    value="percent",
    show_values=True,
    max_label_length=30,
    missing_color="#8e44ad",
    title="Missingness Subsets (ascending)",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=True,
    legend_title="Pattern",
    width=800,
    height=450,
).save()
```
