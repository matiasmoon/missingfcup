# Boxplot

Box or violin plot comparing the distribution of column `x` between rows where `color_by` is present vs. missing.
Diverging distributions suggest MAR or MNAR.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

Do passengers with unknown age pay different fares?

```python
md.boxplot(x="fare", color_by="age").show()
```

## Intermediate

Switch to violin plot and add a title.

```python
md.boxplot(
    x="fare",
    color_by="age",
    title="Fare by age missingness",
    plot_type="violin",
).show()
```

## Advanced

Custom colours, light theme, save.

```python
md.boxplot(
    x="fare",
    color_by="age",
    title="Titanic: Fare Distribution split by age Missingness",
    plot_type="violin",
    missing_color="#e74c3c",
    present_color="#27ae60",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=True,
    legend_title="Age",
    width=700,
    height=450,
).save()
```
