# Dendrogram

Hierarchical clustering of columns by their missingness correlation.
Columns that go missing together end up in the same cluster.

> Requires `scikit-learn` — `pip install scikit-learn`

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from missingfcup import MissingData

data = load_breast_cancer(as_frame=True)
df = data.data.copy()
rng = np.random.default_rng(42)
for col in df.columns:
    idx = rng.choice(len(df), size=int(len(df) * 0.2), replace=False)
    df.loc[idx, col] = np.nan
md = MissingData(df)
```

## Basic

```python
md.dendrogram().show()
```

## Intermediate

Top 15 columns, Ward linkage.

```python
md.dendrogram(
    title="Column Missingness Clustering",
    max_columns=15,
    linkage_method="ward",
).show()
```

## Advanced

Absolute correlation, custom line style, light theme, save.

```python
md.dendrogram(
    title="Column Missingness Clustering — Ward Linkage",
    selected_columns=None,
    max_columns=20,
    drop_constant_columns=True,
    linkage_method="ward",
    use_abs_correlation=True,
    line_width=2,
    line_color="#4C78A8",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=False,
    width=900,
    height=500,
).save()
```
