# Density

Overlapping KDE curves showing how the distribution of column `x` differs between rows where `color_by` is present (green) vs. missing (red).
Diverging distributions are a key signal for MAR.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

Does the fare distribution differ between passengers with known vs. unknown age?

```python
md.density(x="fare", color_by="age").show()
```

## Intermediate

Add a title and increase KDE resolution.

```python
md.density(
    x="fare",
    color_by="age",
    title="Fare Distribution: age present vs. missing",
    n_points=400,
).show()
```

## Advanced

Reduced fill opacity, custom colours, light theme, save.

```python
md.density(
    x="fare",
    color_by="age",
    title="Titanic: Does fare predict age missingness?",
    n_points=400,
    fill_opacity=0.3,
    missing_color="#e74c3c",
    present_color="#27ae60",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=True,
    legend_title="Age",
    width=800,
    height=450,
).save()
```
