# Heatmap

Binary missingness heatmap: each cell is coloured red (missing) or green (present).
Rows can be sorted by any column value to reveal structural patterns.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

```python
md.heatmap().show()
```

## Intermediate

Sort rows by `pclass` to see whether missing `age` and `deck` cluster around passenger class.

```python
md.heatmap(
    title="Titanic: Missingness Heatmap",
    order_by=[{"column": "pclass", "direction": "asc"}],
).show()
```

## Advanced

Sort by fare, highlight the sort column, custom colours, save.

```python
md.heatmap(
    title="Titanic: Missingness Pattern (sorted by fare)",
    selected_columns=["pclass", "age", "sibsp", "parch", "fare", "deck"],
    ignore_high_missingness=False,
    high_missingness_threshold=0.95,
    max_columns=6,
    order_by=[{"column": "fare", "direction": "asc"}],
    show_colorscale=True,
    group_by_mode="binary",
    xgap=2,
    ygap=1,
    max_label_length=20,
    order_by_border_color="#ff7f0e",
    order_by_border_width=3,
    missing_color="#e74c3c",
    present_color="#27ae60",
    background_color="#f5f5f5",
    text_color="#1a1a1a",
    show_legend=True,
    legend_title="Missingness",
    width=950,
    height=600,
).save()
```
