# Heatmap Biserial

Point-biserial correlation between observed column values and missingness indicators.
Each cell [i, j]: does the distribution of column i differ between rows where j is present vs. missing?
Non-zero values are the key signal for MAR.

> Requires `scikit-learn` — `pip install scikit-learn`

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from missingfcup import MissingData

data = load_breast_cancer(as_frame=True)
df = data.data.copy()
rng = np.random.default_rng(42)
for col in df.columns:
    idx = rng.choice(len(df), size=int(len(df) * 0.2), replace=False)
    df.loc[idx, col] = np.nan
md = MissingData(df)
```

## Basic

```python
md.heatmap_biserial().show()
```

## Intermediate

Restrict the row axis to five columns, show numeric values.

```python
md.heatmap_biserial(
    title="Value vs. Missingness Correlation",
    selected_value_columns=[
        "mean radius", "mean area", "mean perimeter",
        "mean concavity", "mean smoothness",
    ],
    show_values=True,
).show()
```

## Advanced

Independent row/column axes, divergent palette, light theme, save.

```python
md.heatmap_biserial(
    title="Value vs. Missingness Association",
    selected_columns=None,
    selected_value_columns=[
        "mean radius", "mean area", "mean perimeter",
        "mean concavity", "mean smoothness",
    ],
    selected_missing_columns=[
        "mean texture", "mean compactness",
        "mean concave points", "mean fractal dimension",
    ],
    colorscale="RdBu",
    show_values=True,
    max_columns=10,
    drop_constant_columns=True,
    order_by_missingness=True,
    order="desc",
    value_round=2,
    show_colorbar=True,
    show_upper_triangle=False,
    nan_color="#eeeeee",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=False,
    width=850,
    height=500,
).save()
```
