# Heatmap Correlation

Heatmap of missingness correlation between columns.
A high positive value means the two columns tend to be missing in the same rows.

> Requires `scikit-learn` — `pip install scikit-learn`

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from missingfcup import MissingData

data = load_breast_cancer(as_frame=True)
df = data.data.copy()
rng = np.random.default_rng(42)
for col in df.columns:
    idx = rng.choice(len(df), size=int(len(df) * 0.2), replace=False)
    df.loc[idx, col] = np.nan
md = MissingData(df)
```

## Basic

```python
md.heatmap_correlation().show()
```

## Intermediate

Top 10 columns, lower triangle only, with value labels.

```python
md.heatmap_correlation(
    title="Missingness Correlation",
    max_columns=10,
    show_values=True,
    show_upper_triangle=False,
).show()
```

## Advanced

Divergent colour scale, dark theme, ordered by missingness, save.

```python
md.heatmap_correlation(
    title="Missingness Correlation — Top 12 Columns",
    selected_columns=None,
    colorscale="RdBu",
    show_values=True,
    max_columns=12,
    drop_constant_columns=True,
    order_by_missingness=True,
    order="desc",
    value_round=2,
    show_colorbar=True,
    show_upper_triangle=False,
    nan_color="#333333",
    background_color="#0b0b0b",
    text_color="#f2f2f2",
    show_legend=False,
    width=750,
    height=600,
).save()
```
