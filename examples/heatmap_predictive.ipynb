# Heatmap Predictive

Heatmap of present-vs-missing correlation: does observing column i predict the missingness of column j?
A high positive value means rows where i is present are more likely to have j missing.

> Requires `scikit-learn` — `pip install scikit-learn`

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from missingfcup import MissingData

data = load_breast_cancer(as_frame=True)
df = data.data.copy()
rng = np.random.default_rng(42)
for col in df.columns:
    idx = rng.choice(len(df), size=int(len(df) * 0.2), replace=False)
    df.loc[idx, col] = np.nan
md = MissingData(df)
```

## Basic

```python
md.heatmap_predictive().show()
```

## Intermediate

Top 10 columns, lower triangle only, with value labels.

```python
md.heatmap_predictive(
    title="Predictive Missingness Correlation",
    max_columns=10,
    show_values=True,
    show_upper_triangle=False,
).show()
```

## Advanced

Divergent colour scale, ordered by missingness, light theme, save.

```python
md.heatmap_predictive(
    title="Predictive Correlation — Top 12 Columns",
    selected_columns=None,
    colorscale="RdBu",
    show_values=True,
    max_columns=12,
    drop_constant_columns=True,
    order_by_missingness=True,
    order="desc",
    value_round=2,
    show_colorbar=True,
    show_upper_triangle=False,
    nan_color="#eeeeee",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=False,
    width=750,
    height=600,
).save()
```
