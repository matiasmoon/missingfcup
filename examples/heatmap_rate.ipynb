# Heatmap Rate

Single-row heatmap where each cell shows the missing rate for one column.
Useful for a compact colour-coded overview of which columns are most incomplete.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

```python
md.heatmap_rate().show()
```

## Intermediate

Red colour scale, sorted descending, values shown.

```python
md.heatmap_rate(
    title="Missing Rate per Column",
    colorscale="Reds",
    order="desc",
    show_values=True,
).show()
```

## Advanced

Percentage scale, yellow-orange-red ramp, save.

```python
md.heatmap_rate(
    title="Missing Rate — All Columns",
    selected_columns=None,
    scale="percentage",
    colorscale="YlOrRd",
    show_values=True,
    max_columns=15,
    order_by_missingness=True,
    order="desc",
    value_round=1,
    show_colorbar=True,
    max_label_length=20,
    max_labels_with_values=20,
    background_color="#fafafa",
    text_color="#111111",
    show_legend=True,
    legend_title="Rate",
    width=1000,
    height=220,
).save()
```
