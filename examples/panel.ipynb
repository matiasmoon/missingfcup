# Panel

Combine multiple plot objects into a single multi-panel figure with a shared title and automatic grid layout.
Accepts up to 6 plots in a grid of at most `max_cols` columns.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData, Panel

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

Two plots side by side.

```python
Panel(
    plots=[
        md.barchart_count(title="Missing Count"),
        md.barchart_rate(title="Missing Rate"),
    ],
    title="Titanic: Missingness Overview",
).show()
```

## Intermediate

Three plots: count, rate heatmap, and dataset totals.

```python
Panel(
    plots=[
        md.barchart_count(title="Count"),
        md.barchart_rate(title="Rate"),
        md.barchart_total(title="Total"),
    ],
    title="Titanic: Missingness Overview",
).show()
```

## Advanced

Six plots in a 3-column dark-theme grid, save.

```python
panel = Panel(
    plots=[
        md.barchart_count(title="Missing Count"),
        md.barchart_rate(title="Missing Rate"),
        md.barchart_total(title="Dataset Total"),
        md.heatmap_rate(title="Rate Heatmap"),
        md.scatterplot(x="fare", y="age", title="Fare vs. Age"),
        md.density(x="fare", color_by="age", title="Fare by age missingness"),
    ],
    title="Titanic: Full Missingness Overview",
    description="Missing-data diagnostics across 891 passengers",
    max_cols=3,
    max_plots=6,
    background_color="#0b0b0b",
    text_color="#f2f2f2",
    width=1400,
    height=750,
)
panel.show()
panel.save()
```
