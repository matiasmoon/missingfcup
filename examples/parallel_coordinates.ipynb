# Parallel Coordinates

Each line is one row, each axis one column.
Lines are coloured by whether a chosen column is present (green) or missing (red).
Useful for spotting if rows with a missing value form a distinct subgroup in feature space.

> Requires `scikit-learn` — `pip install scikit-learn`

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from missingfcup import MissingData

data = load_breast_cancer(as_frame=True)
df = data.data.copy()
rng = np.random.default_rng(42)
for col in df.columns:
    idx = rng.choice(len(df), size=int(len(df) * 0.2), replace=False)
    df.loc[idx, col] = np.nan
md = MissingData(df)
```

## Basic

```python
md.parallel_coordinates().show()
```

## Intermediate

Restrict to 6 columns and colour by `mean texture` missingness.

```python
md.parallel_coordinates(
    title="Parallel Coordinates — coloured by mean texture missingness",
    selected_columns=[
        "mean radius", "mean texture", "mean perimeter",
        "mean area", "mean smoothness", "mean compactness",
    ],
    missingness_color_column="mean texture",
).show()
```

## Advanced

Reduce line opacity and width, expand column set, light theme, save.

```python
md.parallel_coordinates(
    title="Feature Space coloured by mean texture Missingness",
    selected_columns=[
        "mean radius", "mean texture", "mean perimeter",
        "mean area", "mean smoothness", "mean compactness",
        "mean concavity", "mean concave points",
    ],
    missingness_color_column="mean texture",
    line_opacity=0.25,
    line_width=0.7,
    missingness_only=False,
    normalize=True,
    missing_color="#e74c3c",
    present_color="#27ae60",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=True,
    legend_title="Missingness",
    width=1000,
    height=450,
).save()
```
