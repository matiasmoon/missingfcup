# Scatterplot

Scatter plot that keeps missing values visible by offsetting them to fixed positions at the axis margins.
Points can be coloured by the missingness of a third column.

> Requires `seaborn` — `pip install seaborn`

```python
import seaborn as sns
from missingfcup import MissingData

df = sns.load_dataset("titanic")
md = MissingData(df)
```

## Basic

`age` has ~20% NaN — those rows appear along the bottom margin.

```python
md.scatterplot(x="fare", y="age").show()
```

## Intermediate

Colour points by `deck` missingness and increase point size.

```python
md.scatterplot(
    x="fare",
    y="age",
    title="Fare vs. Age — coloured by deck missingness",
    missingness_color_column="deck",
    point_size=7,
).show()
```

## Advanced

Tune axis padding, missing-value jitter, custom colours, light theme, save.

```python
md.scatterplot(
    x="fare",
    y="age",
    title="Titanic: Fare vs. Age",
    missingness_color_column="deck",
    point_size=8,
    axis_padding=0.12,
    point_opacity=0.7,
    jitter=0.02,
    missing_jitter=0.3,
    jitter_seed=42,
    xaxis_range=None,
    yaxis_range=None,
    missing_color="#e74c3c",
    present_color="#27ae60",
    background_color="#fafafa",
    text_color="#111111",
    show_legend=True,
    legend_title="Deck",
    width=800,
    height=500,
).save()
```
