```python
import pandas as pd
import numpy as np
from ucimlrepo import fetch_ucirepo
from missingfcup import MissingData
from mdatagen.univariate.uMCAR import uMCAR
from mdatagen.univariate.uMAR  import uMAR
from mdatagen.univariate.uMNAR import uMNAR

# ── Cargar dataset ──────────────────────────────────────────
student_performance = fetch_ucirepo(id=320)
X_raw = student_performance.data.features
y = student_performance.data.targets.iloc[:, 0].to_numpy()
```

```python
# Encoding
X = X_raw.copy()
cat_cols = X.select_dtypes(include=["object", "category"]).columns
for col in cat_cols:
    X[col] = pd.factorize(X[col])[0]

df = pd.concat([X, pd.Series(y, name="G1")], axis=1)
md_complete = MissingData(df)
```

```python

# ── MCAR univariado ─────────────────────────────────────────
u_mcar = uMCAR(X=X, y=y, missing_rate=20, x_miss="absences")
data_u_mcar = u_mcar.random()
md_mcar = MissingData(data_u_mcar)
```

```python

# ── MAR univariado ──────────────────────────────────────────
u_mar = uMAR(X=X, y=y, missing_rate=20, x_obs="absences", x_miss="studytime")
data_u_mar = u_mar.highest()  # ← cualquiera de los 5, todos son MAR
md_mar = MissingData(data_u_mar)
```

```python

# ── MNAR univariado ─────────────────────────────────────────
u_mnar = uMNAR(X=X, y=y, missing_rate=20, x_miss="absences", threshold=1)
data_u_mnar = u_mnar.run()
md_mnar = MissingData(data_u_mnar)
```

```python
# MCAR: missingness in 'absences' is purely random — no pattern with any other variable
md_mcar.scatterplot(x="absences", y="studytime", width=400, height=400).show()

# MAR: missingness in 'studytime' is explained by high values of 'absences' (x_obs).
# Using missingness_color_column reveals the mechanism directly:
# both axes are always observed in this dataset, so all points sit in the plot area —
# color (red = missing studytime) concentrates on the right (high absences) side.
md_mar.scatterplot(
    x="absences",
    y="freetime",
    missingness_color_column="studytime",
    width=400,
    height=400,
).show()

# MNAR: missingness in 'absences' is driven by the value of 'absences' itself (highest 20%)
md_mnar.scatterplot(x="absences", y="studytime", width=400, height=400).show()
```
