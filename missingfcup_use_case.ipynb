# missingfcup: Simple Missing-Data Use Case

This notebook loads a small dataset and showcases all missing-data plots in `missingfcup`, plus a couple of quick analytics checks.

## 1. Set Up: Install and Import Dependencies

```python
# If running in a clean environment, uncomment these lines.
# %pip install -q -e .

import numpy as np
import pandas as pd

from missingfcup import MissingData, Panel
```

## 2. Load Sample Data for the Use Case

We use the classic Iris dataset (UCI), served via the seaborn GitHub mirror.

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

df = pd.read_csv(url)

# Inject a small amount of missingness for demonstration
rng = np.random.default_rng(7)
mask = rng.random(df.shape) < 0.03
df = df.mask(mask)

df.head()
```

## 3. Explore Missingness and Plots

```python
md = MissingData(df)

md.column_missing_rate.sort_values(ascending=False).head()
```

```python
# All plots from MissingData
md.heatmap().show()
md.missing_count_barchart().show()
md.column_missing_rate_heatmap().show()
md.missingness_correlation_heatmap().show()
md.pattern_barchart().show()
md.scatterplot(x="sepal_length", y="sepal_width").show()
```

```python
# Combine a few plots in a single panel
panel = Panel([
    md.heatmap(title="Missingness"),
    md.missing_count_barchart(title="Counts"),
    md.column_missing_rate_heatmap(title="Rates"),
])

panel.show()
```

```python
# Quick analytics checks
print(md.column_missing_rate.sort_values(ascending=False).head())
print(md.rows_above_missing_threshold(0.2))
print(md.columns_above_missing_threshold(0.2))
print(md.missing_pattern_counts(max_patterns=5))
print(md.missing_correlation().head())
print(md.perfectly_correlated_missing_columns())
```
